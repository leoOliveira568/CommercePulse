```python
from pathlib import Path

import numpy as np
import pandas as pd

pd.set_option("display.max_columns", 100)
pd.set_option("display.width", 120)

BASE_DIR = Path("..")
RAW_DIR = BASE_DIR / "data" / "raw"
PROCESSED_DIR = BASE_DIR / "data" / "processed"

PROCESSED_DIR.mkdir(parents=True, exist_ok=True)
```

```python
orders = pd.read_csv(RAW_DIR / "olist_orders_dataset.csv")
order_items = pd.read_csv(RAW_DIR / "olist_order_items_dataset.csv")
payments = pd.read_csv(RAW_DIR / "olist_order_payments_dataset.csv")
reviews = pd.read_csv(RAW_DIR / "olist_order_reviews_dataset.csv")
customers = pd.read_csv(RAW_DIR / "olist_customers_dataset.csv")
sellers = pd.read_csv(RAW_DIR / "olist_sellers_dataset.csv")
products = pd.read_csv(RAW_DIR / "olist_products_dataset.csv")
category_translation = pd.read_csv(RAW_DIR / "product_category_name_translation.csv")
```

```python
datasets = {
    "orders": orders,
    "order_items": order_items,
    "payments": payments,
    "reviews": reviews,
    "customers": customers,
    "sellers": sellers,
    "products": products,
    "category_translation": category_translation,
}

summary = []

for name, df in datasets.items():
    summary.append({
        "dataset": name,
        "rows": df.shape[0],
        "columns": df.shape[1],
        "duplicated_rows": df.duplicated().sum(),
        "missing_cells": df.isna().sum().sum(),
    })

pd.DataFrame(summary).sort_values("rows", ascending=False)
```

```python
for name, df in datasets.items():
    print(f"\n{name.upper()}")
    print(df.columns.tolist())
```

```python
df = pd.read_csv(PROCESSED_DIR / "commercepulse_orders_items.csv")

df.head()
```

```python
kpis = {
    "total_orders": df["order_id"].nunique(),
    "total_items_sold": len(df),
    "total_revenue": df["item_revenue"].sum(),
    "average_ticket": df.groupby("order_id")["item_revenue"].sum().mean(),
    "average_review_score": df["review_score"].mean(),
    "late_delivery_rate": df["is_late"].mean(),
    "average_delivery_days": df["delivery_days"].mean(),
    "unique_customers": df["customer_unique_id"].nunique(),
    "unique_sellers": df["seller_id"].nunique(),
    "unique_categories": df["product_category"].nunique(),
}

pd.DataFrame(kpis.items(), columns=["metric", "value"])
```
